```python
# Dependencias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

En este caso el set de datos se obtuvo directamente de la pagina de la [Agencia Nacional de Transito del Ecuador](https://www.ant.gob.ec/visor-de-siniestralidad-estadisticas/) la cual tiene datos desde enero de 2017 a abril de 2023.

```python
df = pd.read_csv("../data/BDD_ENE_2017_ABR_2023.csv", delimiter=";", low_memory=False)
df
```

## Analisis Estadistico Fallecidos y Lesionados por año

```python
df['FALLECIDOS'].describe()
```

```python
df['LESIONADOS'].describe()
```

```python
data = df[['ANIO', 'LESIONADOS', 'FALLECIDOS']]
```

```python
plt.scatter(x=data['ANIO'], y=data['LESIONADOS'], c='blue', label='LESIONADOS')
plt.scatter(x=data['ANIO'], y=data['FALLECIDOS'], c='red', label='FALLECIDOS')
plt.xlabel('ANIO')
plt.ylabel('NUMERO DE LECIONADOS Y FALLECIDOS')
plt.title('Gráfico de Dispersión')
plt.legend()
plt.show()
```

```python
sns.histplot(data=data, x='ANIO', discrete=True)
# Agregar etiquetas
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Gráfico de Frecuencias')

# Mostrar el gráfico de frecuencias
plt.show()
```

Notamos que hay una ligera disminucion despues del 2017 a medida que nos hacercamos al 2020, puesto que en el 2020 hubo restricciones de movilidad, podemos pensar que debido a esto la tasa de accidentes baja considerablemente.

## Anlasis por Provincias

```python
prov = pd.pivot_table(df, values=['FALLECIDOS', 'LESIONADOS'], index='PROVINCIA', aggfunc=np.sum)
prov
```

```python
plt.plot(prov)
plt.xticks(rotation=90)
plt.show()
```

Se observa un gran numero de lesionados en la provincia del Guayas y Pichincha, esto se debe a que son dos de las provincias más habitadas del pais, y por ende hay mayor trafico de vehiculos.

```python
hxprov = pd.pivot_table(df, values='PERIODO_2', index='PROVINCIA')
hxprov
```

Notamos que en promedio, la mayoria de los accidentes se dan en horario del medio día.
