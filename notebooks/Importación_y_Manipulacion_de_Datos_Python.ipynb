```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

(a) Crea una matriz de 10 números enteros aleatorios entre 1 y 100 y, a continuación, calcula la media y la desviación típica de la matriz.

```python
a = np.random.randint(1, high=101, size=10)
print(a)
print(a.mean())
print(a.std())
```

(b) Crea una matriz bidimensional de 3 filas y 4 columnas con valores enteros aleatorios. A continuación, calcula la suma de cada fila y columna.

```python
b = np.random.randint(10, size=(3, 4))
print("2D Array:\n", b)

row_sum = np.sum(b, axis=1)
print("Row Sum:\n", row_sum)

col_sum = np.sum(b, axis=0)
print("Column Sum:\n", col_sum)
```

(c) Crea un DataFrame con 3 columnas: "Nombre", "Edad" y "Ciudad". Añade al menos 5 filas de datos al DataFrame. A continuación, filtra el DataFrame para que sólo incluya filas en las que la edad de la persona sea mayor o igual a 30 años.

```python
# Crear el DataFrame
data = {
    'Nombre': ['Juan', 'María', 'Pedro', 'Ana', 'Luis'],
    'Edad': [25, 35, 42, 29, 31],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Bilbao']
}

df = pd.DataFrame(data)

# Filtrar el DataFrame para incluir solo personas mayores de 30 años
df_filtrado = df[df['Edad'] >= 30]

print(df_filtrado)
```

d) Cargue un archivo CSV en un DataFrame y, a continuación, calcule la media, la mediana y la moda de una de las columnas del DataFrame.

```python
df = pd.read_csv('ServiceArea.csv')
df
```

```python
# Calcular la media
media = df['County'].mean()

# Calcular la mediana
mediana = df['County'].median()

# Calcular la moda
moda = df['County'].mode()[0]

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)
```

(e) Crea un diagrama de dispersión con valores aleatorios de x e y comprendidos entre 1 y 100.

```python
np.random.seed(1)
x = np.random.randint(1, 101, size=50)
y = np.random.randint(1, 101, size=50)

# Create scatter plot
plt.scatter(x, y)

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')

# Show the plot
plt.show()
```
