(g) Crea una matriz de 10 números enteros aleatorios entre 1 y 100 y, a continuación, calcula la media y la desviación típica de la matriz.

```r
# Crear la matriz con números aleatorios
matriz <- sample(1:100, 10, replace = TRUE)

# Calcular la media
media <- mean(matriz)

# Calcular la desviación estándar
desviacion <- sd(matriz)

# Imprimir los resultados
print(matriz)
print(paste("Media:", media))
print(paste("Desviación estándar:", desviacion))
```

(h) Crea una matriz bidimensional de 3 filas y 4 columnas con valores enteros aleatorios. A continuación, calcula la suma de cada fila y columna.

```r
# Crear la matriz bidimensional con valores aleatorios
matriz2 <- matrix(sample(1:100, 12, replace = TRUE), nrow = 3)

# Calcular la suma de cada fila
sumas_filas <- rowSums(matriz2)

# Calcular la suma de cada columna
sumas_columnas <- colSums(matriz2)

# Imprimir los resultados
print(matriz2)
print("Suma de cada fila:")
print(sumas_filas)
print("Suma de cada columna:")
print(sumas_columnas)
```

(k) Cargue un archivo CSV en un DataFrame y, a continuación, calcule la media, la mediana y la moda de una de las columnas del DataFrame.

```r
# Cargar el archivo CSV en un DataFrame
df <- read.csv("ServiceArea.csv")
head(df, 5)
```

```r
# Definir función para calcular la moda
Mode <- function(x) {
  ux <- unique(x)
  ux[which.max(tabulate(match(x, ux)))]
}

# Calcular la media de una columna
media <- mean(df$IssuerId)

# Calcular la mediana de una columna
mediana <- median(df$IssuerId)

# Calcular la moda de una columna
moda <- Mode(df$IssuerId)

# Imprimir los resultados
print(paste("Media:", media))
print(paste("Mediana:", mediana))
print(paste("Moda:", moda))
```

(l) Crea un diagrama de dispersión con valores aleatorios de x e y comprendidos entre 1 y 100.

```r
# Generar valores aleatorios de x e y
set.seed(42)  # Para reproducibilidad
x <- runif(100, min = 1, max = 100)
y <- runif(100, min = 1, max = 100)

# Crear el diagrama de dispersión
plot(x, y, main = "Diagrama de dispersión", xlab = "X", ylab = "Y", pch = 16, col = "blue")
```
